```python
import yfinance as yf
msft = yf.Ticker("MSFT")
msft.info
```

```python
print(type(msft))
```

```python
hist = msft.history(start='1990-01-01')
```

```python
print(hist[''])
```

```python
bucks = 1000000.0
stocks = 0
sumHLBefore = 0
for i in range(len(hist)):
    if (hist['Open'][i] < hist['Close'][i]):
        if  (hist['High'][i] + hist['Low'][i] > sumHLBefore) and stocks == 0:
            stocks += int(bucks/max(((hist['High'][i] + hist['Low'][i])/2), hist['Open'][i]))
            bucks -= stocks*(hist['High'][i] + hist['Low'][i])/2
        sumHLBefore = hist['High'][i] + hist['Low'][i]
    
    if (hist['Open'][i] >= hist['Close'][i]) and stocks != 0:
        bucks += stocks*min(((hist['High'][i] + hist['Low'][i])/2), hist['Open'][i])
        stocks = 0
    
    print(i, bucks, stocks)
```
